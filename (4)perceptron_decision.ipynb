```python
import numpy as np
import matplotlib.pyplot as plt

# Sample input features and binary labels
X = np.array([[0.5, 1.2], [1.1, 2.5], [1.0, 0.8], [0.2, 1.0], [1.5, 1.8]])
y = np.array([[0], [1], [1], [0], [1]])

# Initialize weights and bias
w = np.zeros(2)
b = 0
learning_rate = 0.1
epochs = 10

# Perceptron training loop
for _ in range(epochs):
    for i in range(len(X)):
        y_pred = np.dot(w, X[i]) + b
        y_pred = 1 if y_pred > 0 else 0
        error = y[i][0] - y_pred
        w += learning_rate * error * X[i]
        b += learning_rate * error

# Calculate decision boundary line
x1_values = np.linspace(0, 2, 10)
x2_values = -(w[0] * x1_values / w[1]) - (b / w[1])

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=y[:, 0], cmap='bwr', edgecolors='k', marker='o')
plt.plot(x1_values, x2_values, 'k-', label='Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.show()

# Print final weights and bias
print(f"Final weights: {w}")
print(f"Final bias: {b}")
```
