```python
!pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)

plt.plot(x, linear(x), label='Linear', color='blue')
plt.title("Linear")
plt.grid(True)

plt.plot(x, sigmoid(x), label="Sigmoid", color='green')
plt.title("Sigmoid")
plt.grid(True)

plt.plot(x, tanh(x), label='Tanh', color='yellow')
plt.title("Tanh")
plt.grid(True)

plt.plot(x, relu(x), label='ReLU', color='red')
plt.title("ReLU")
plt.grid(True)

plt.plot(x,leaky_relu(x),label='Leaky ReLU',color='orange')
plt.title("Leaky_ReLU")
plt.grid(True)

plt.figure(figsize=(10, 6))
plt.plot(x, linear(x), label='Linear', color='blue')
plt.plot(x, sigmoid(x), label='Sigmoid', color='green')
plt.plot(x, tanh(x), label='Tanh', color='purple')
plt.plot(x, relu(x), label='ReLU', color='red')
plt.plot(x, leaky_relu(x), label='Leaky ReLU', color='orange')
plt.xlabel("Input (x)")
plt.ylabel("Activation Output")
plt.title("All Activation Functions")
plt.legend()
plt.grid(True)

plt.show()
```
