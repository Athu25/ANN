```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

def create_cnn_model(input_shape=(64, 64, 3)):
    input_layer = layers.Input(shape=input_shape)
    
    x = layers.Conv2D(32, (3, 3), activation='relu')(input_layer)
    x = layers.MaxPooling2D((2, 2))(x)
    
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)

    bbox_output = layers.Dense(4, activation='sigmoid', name='bbox_output')(x)   # Bounding box
    class_output = layers.Dense(1, activation='sigmoid', name='class_output')(x) # Object detection

    model = models.Model(inputs=input_layer, outputs=[bbox_output, class_output])

    model.compile(
        optimizer='adam',
        loss={'bbox_output': 'mse', 'class_output': 'binary_crossentropy'},
        metrics={'bbox_output': 'accuracy', 'class_output': 'accuracy'}
    )

    return model

# Dummy training data
X_train = np.random.rand(100, 64, 64, 3)             # 100 images of size 64x64x3
y_train_bbox = np.random.rand(100, 4)                # 100 bounding boxes (x, y, w, h)
y_train_class = np.random.randint(0, 2, size=(100, 1))  # Binary labels (0 or 1)

# Create and train the model
model = create_cnn_model()
model.fit(X_train, {'bbox_output': y_train_bbox, 'class_output': y_train_class}, epochs=10)

# Test on a new image
test_image = np.random.rand(1, 64, 64, 3)
pred_bbox, pred_class = model.predict(test_image)

# Output predictions
print("Predicted Bounding Box:", pred_bbox)
print("Predicted Class (Object detected?):", pred_class)

# Visualize the result
plt.imshow(test_image[0])
plt.title(f"Object Detected: {'Yes' if pred_class > 0.5 else 'No'}")
plt.show()
```
